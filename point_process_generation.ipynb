```python
# Making the waveform of firing rate
# ----------------------------------

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

time = np.linspace(0, 5, 5001)
firing_rate = 3*np.sin(2*np.pi*time) + 1
firing_rate[firing_rate < 0] = 0

plt.plot(time, firing_rate)
plt.xlabel('time [s]')
plt.ylabel('firing rate [Hz]')
```

```python
# Generating spike trains
# -----------------------

n_trial = 1000
r = np.random.rand(n_trial, len(time))

firing_times = []
for i in range(n_trial):
    for j in range(len(time)):
        if r[i, j] < firing_rate[j]*1e-3:
            firing_times.append([time[j], i])

firing_times = np.array(firing_times)
plt.plot(firing_times[:, 0], firing_times[:, 1], 'o', markersize=1)
plt.xlabel('time [s]')
plt.ylabel('trial #')
```
